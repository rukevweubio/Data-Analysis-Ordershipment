```python
%load_ext sql
engine = create_engine('sqlite:///databasedesignfor.db')
%config SQL_conn.name='engine'
connection =sqlite3.connect('databasedesignfor.db')
cursor=connection.cursor()
%sql sqlite:///databasedesignfor.db
```

```python
cursor.execute('''
    create table if not  exists accounts(
    Accountnumber int not null   primary key,
    AccountType varchar(50) not null,
    Balance  decimal(10,2) not null,
    Dateopened date not null,
    CustomerId int not null,
    foreign key(CustomerId) references customers(CustomerId)
    )
''')
cursor.execute('''
    create  table if  not exists branches(
    BranchId int not null    primary key,
    BranchName varchar(50) not null,
    BranchAddress varchar(200) not null,
    BranchPhoneNumber varchar(50) not null)


''')
cursor.execute('''
    create table if not exists customers(
    CustomerId int not null  primary key,
    Firstname varchar(50) not null,
    Lastname  varchar(50) not null,
    DateofBirth date not null,
    Address  varchar(50)  not null,
    Phonenumber varchar(50),
    City varchar(50 ) not null

    )
''')
cursor.execute('''
    create table if not exists employee(
    EmployeeID int primary key,
    Name varchar(20) not null,
    Position varchar(50) not null,
    BranchID int not null,
    HireDate date not null,
    foreign key (BranchID) references Branch(BranchID))
''')

cursor.execute('''
    create table if not exists Loan(
    LoanID int primary key,
    CustomerID int not null,
    LoanAmount Decimal(10,2) not null,
    LoanType VArchar(20) not null,
    IssueDate date not null,
    InterestRate decimal(10,2) not null,
    foreign key(CustomerID) references customers(customerID)
)
''')
cursor.execute('''
    create table if not exists transactions(
    TransactionID int primary key,
    AccountNumber int not null,
    TransactionType varchar(50) not null,
    Amount decimal(10,2) not null,
    TransactionDate date not null
    )
''')

```

How many customers are there in each branch?

```python
%%sql
select b.BranchName, count(*), count(customerid) from customers c join branches b on  c.customerid = b.branchid
group by b.BranchName
```

What is the total loan amount taken by each customer?

```python
%%sql
select firstname, lastname, sum(loanAmount) from customers c join loan l on c.customerid= l.customerid
group by firstname, lastname
order by sum(loanAmount) desc limit (10)
```

bottom last customer with the minimum amount of loan borrow

```python
%%sql
select firstname, lastname, sum(loanAmount) from customers c join loan l on c.customerid= l.customerid
group by firstname, lastname
order by sum(loanAmount) asc limit (10)
```

What is the average loan amount per branch?

```python
%%sql
select b.branchname, avg(loanAmount) as aveargeAmount from  loan l join branches b on(l.loanid= b.branchid)
group by b.branchname
order by avg(loanAmount)   desc
```

loan type with the higest amount borrow

```python
%%sql
select  loanType, round((sum(loanAmount)/count(*)),3) as amount from loan
group by loanType
```

Which customers have loans exceeding a certain amount (e.g., $70,000)

```python
%%sql
select firstname, lastname, sum(loanAmount)as amount from customers c join loan l on c.customerid= l.customerid
group by firstname, lastname
having sum(loanAmount)> 70000;
```

What is the total amount of transactions for each account?

```python
%%sql
select  accountType,count(*), round(sum(amount),3) from accounts  a join transactions t on (a.accountnumber=t.accountnumber)
group by accountType
```

What are the top 5 transactions in terms of amount?

```python
%%sql
select  c.firstname, c.lastname, count(*), sum(amount) from accounts  a join transactions t on (a.accountnumber=t.accountnumber)
join customers c on (c.customerid=a.customerid)
group by c.firstname,c.lastname
order by count(*) desc limit 5
```

What are the bottom 5 transactions in terms of amount

```python
%%sql
select  c.firstname, c.lastname, count(*), sum(amount) from accounts  a join transactions t on (a.accountnumber=t.accountnumber)
join customers c on (c.customerid=a.customerid)
group by c.firstname,c.lastname
order by count(*) asc limit 5
```

How many accounts does each customer have
